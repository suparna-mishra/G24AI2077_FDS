**Question 1 : MapReduce Program for Cruise Aggregation problem**
1. It takes the cruise.csv as input and splits the line by comma(,)  .
2. Skips the header .
3. For each cruise_line value the code calculates the total no of ships it has, average tonnage(rounded off to 2 decimal places) and maximum crew size .

```python
%%writefile Q1_CruiseAgg.py

from mrjob.job import MRJob
from mrjob.step import MRStep

class Q1_CruiseAgg(MRJob):

    def mapper(self, _, line):

        if line.startswith('Ship_name'):
            return


        fields = line.split(',')


        cruise_line = fields[1]
        try:
            tonnage = float(fields[3])
            crew = float(fields[8])
        except (ValueError, IndexError):
            return


        yield cruise_line, (1, tonnage, crew)

    def combiner(self, cruise_line, values):
        total_ships = 0
        total_tonnage = 0.0
        max_crew = 0.0

        for count, tonnage, crew in values:
            total_ships += count
            total_tonnage += tonnage
            if crew > max_crew:
                max_crew = crew

        yield cruise_line, (total_ships, total_tonnage, max_crew)

    def reducer(self, cruise_line, values):
        total_ships = 0
        total_tonnage = 0.0
        max_crew = 0.0

        for count, tonnage, crew in values:
            total_ships += count
            total_tonnage += tonnage
            if crew > max_crew:
                max_crew = crew


        avg_tonnage = round(total_tonnage / total_ships, 2)

        yield cruise_line, (total_ships, avg_tonnage, max_crew)

    def steps(self):
        return [
            MRStep(mapper=self.mapper,
                   combiner=self.combiner,
                   reducer=self.reducer)
        ]

if __name__ == '__main__':
    Q1_CruiseAgg.run()
```

**Question 2  : Mapper Code for Company churn rate problem**


1. 1st reads VIP_companies.txt from distributed cache which contains some sample company names extracted from customer_churn.csv
2. Looks up Company names listed in VIP_companies.txt with customer_churn.csv  
3. Counts total customers per company
4. Counts total churned customers per company if churn == 1



```python
%%writefile Q2_Mapper.py
#!/usr/bin/env python3
import sys

# Read VIP companies from Distributed Cache
VIP_companies = set()
with open('VIP_companies.txt', 'r') as f:
    for line in f:
        VIP_companies.add(line.strip())

for line in sys.stdin:
    data = line.strip().split(',')


    company = data[-2]
    churn = data[-1]



    # Only process VIP companies
    if company in VIP_companies:
        # Emit (Company, TOTAL)
        print(f"{company}\tTOTAL\t1")

        # Emit (Company, CHURNED) if applicable
        if churn == '1':
            print(f"{company}\tCHURNED\t1")


```

**Question 2 : Reducer Code for Company churn rate problem**

1. Gets the Mapper output and sums total and churned customers per company
2. Calculates churn rate and rounds up to 4 decimal points


```python
%%writefile Q2_Reducer.py
#!/usr/bin/env python3
import sys
from collections import defaultdict

total_counts = defaultdict(int)
churned_counts = defaultdict(int)

for line in sys.stdin:
    company, metric, count = line.strip().split('\t')
    count = int(count)

    if metric == 'TOTAL':
        total_counts[company] += count
    elif metric == 'CHURNED':
        churned_counts[company] += count

for company in total_counts:
    churn_rate = round(churned_counts[company] / total_counts[company], 4)
    print(f"Company :{company}\t Churn Rate :{churn_rate:.4f}")

```

**Question 3 : Mapper Code for State wise spending problem**

1. Reads e-com_customer.csv and extract the two-letter state code from the Address field using Regular Expression
2. Captures Yearly Amount Spent for each of those states


```python
%%writefile Q3_Mapper.py
#!/usr/bin/env python3
import sys
import csv
import re

reader = csv.reader(sys.stdin)

header = next(reader)
address_ind = header.index('Address')
amount_ind = header.index('Yearly Amount Spent')

for row in reader:
    try:
        address = row[address_ind]
        amount = float(row[amount_ind])

        match = re.search(r',\s*([A-Z]{2})\s*\d{5}', address)
        if match:
            state = match.group(1)
            print(f"{state}\t{amount}")
    except:
        continue
```

**Question 3 : Reducer Code for State wise spending problem**

1. Sums up Yearly Amount Spent value for each state
2. Outputs top 5 most States in terms of total spending


```python
%%writefile Q3_Reducer.py
#!/usr/bin/env python3
import sys


total_spend = 0
state_cnt = {}

for line in sys.stdin:
    state, amount = line.strip().split('\t')
    amount = float(amount)
    if state not in state_cnt:
        state_cnt[state] = 0
    state_cnt[state] += amount



print(f"Top 5 states based on total spending :")
top_states = sorted(state_cnt.items(), key=lambda x: x[1], reverse=True)[:5]

for state, total_spend in top_states:
    print(f"{state}\t{total_spend:.2f}")
```

**Question 4 : Two-step  Ship Filter & Median Length Step on cruise.csv dataset**
1. Filter ships with passenger density > 35.0.
2. Compute the median of the lengths per Cruise line.

```python
%%writefile Q4_MRJob.py
#!/usr/bin/env python3

from mrjob.job import MRJob
from mrjob.step import MRStep
import csv
from statistics import median

class Q4_MRJob(MRJob):

    def steps(self):
        return [
            MRStep(mapper=self.mapper_filter_cruiseline,
                   reducer=self.reducer_collect_lengths),
            MRStep(reducer=self.reducer_compute_median)
        ]

    def mapper_filter_cruiseline(self, _, line):

        try:
            row = next(csv.reader([line]))
            if row[0] == "Ship_name":
                return

            cruise_line = row[1]
            length = float(row[5])
            passenger_density = float(row[7])

            if passenger_density > 35.00:
                yield cruise_line, length
        except:
            pass

    def reducer_collect_lengths(self, cruise_line, length):
        yield cruise_line, list(length)

    def reducer_compute_median(self, cruise_line, list_length):

        all_list_lengths = []
        for lst in list_length:
            all_list_lengths.extend(lst)

        median_length_cruise_line = median(all_list_lengths)


        yield cruise_line, f"{median_length_cruise_line:.2f}"

if __name__ == '__main__':
    Q4_MRJob.run()
```
